匿名函数lambda
语法结构：
lambda 参数1, 参数2, ... : 表达式
冒号后面只能是表达式，表达式的求值结果自动作为返回值，不用写 return。

```python
# def add(x,y):
#    return x+y

add = lambda x,y:x+y
print(add(2,3))
```

```python
nums = [1,2,3,4,5]
squares = list(map(lambda x:x**2,nums))
print(squares)
```

```python
# 1. 配合 map：对每个元素应用函数
nums = [1, 2, 3]
squares = list(map(lambda x: x ** 2, nums))   # [1, 4, 9]
print(squares)
```

```python
# 2. 配合 filter：过滤
evens = list(filter(lambda x: x % 2 == 1, nums))
print(evens)  # [2]
```

```python
# 3. 配合 sorted：自定义排序键
pairs = [(1, 'b'), (3, 'a'), (2, 'c')]
pairs.sort(key=lambda p: p[0])   # 按第二个元素排序 -> [(3,'a'),(1,'b'),(2,'c')]
print(pairs)
```
