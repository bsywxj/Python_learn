enumerate 是什么:
enumerate() 是内置函数，给可迭代对象（列表、字符串、元组等）加上索引下标，把每个元素变成 (索引, 元素) 的元组。
最常用的场景：遍历列表时同时拿到"第几个"和"元素本身"，省得自己维护一个计数器 i = 0。

```python
words = ['apple', 'banana', 'cherry']

for i, w in enumerate(words):
    print(i, w)
```

```python
#加 start= 参数：
for i, w in enumerate(words, start=1):
    print(i, w)
# 1 apple
# 2 banana
# 3 cherry
```

```python
e = enumerate(words)
print(list(e))   # [(0,'apple'),...]
print(list(e))   # []  ← 第二次为空，已耗尽
```

```python
#配合 list comprehension + 下标
# 给每个单词生成 "编号-单词" 字符串
labels = [f"{i}:{w}" for i, w in enumerate(words, 1)]
print(labels)   # ['1:apple', '2:banana', '3:cherry']
```

```python
# 找第一个长度 > 5 的单词下标
idx = next(i for i, w in enumerate(words) if len(w) > 5)
print(idx)   # 1  (banana)
```
